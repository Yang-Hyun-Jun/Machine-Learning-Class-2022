# K-means clustering

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
```

## load data

```python
fname_data = 'assignment_11_data.csv'

feature = np.genfromtxt(fname_data, delimiter=',')

x = feature[:,0]
y = feature[:,1]

number_data     = np.size(feature, 0)
number_feature  = np.size(feature, 1)

print('number of data : {}'.format(number_data))
print('number of feature : {}'.format(number_feature))
```

## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    center_ = np.tile(center, (feature.shape[0], 1))
    distance = np.sqrt((feature[:,0] - center_[:,0]) ** 2 + (feature[:,1] - center_[:,1]) ** 2)

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks

    cluster = feature[label_feature == label]      
    centroid = np.array([np.mean(cluster[:,0]), np.mean(cluster[:,1])])

    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    label_feature = np.zeros((distance.shape[0]))
    for i in range(distance.shape[0]):
      label_feature[i] = np.argmin(distance[i])
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 300     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks

for i in range(number_iteration): 

  centroid_0 = compute_centroid(feature, label_feature_02, 0)
  centroid_1 = compute_centroid(feature, label_feature_02, 1)
  distance_0 = compute_distance(feature, centroid_0).reshape(-1,1)
  distance_1 = compute_distance(feature, centroid_1).reshape(-1,1)
  distance_all = np.concatenate([distance_0, distance_1], axis=1)

  centroid_iteration_02[i,0,:] = centroid_0
  centroid_iteration_02[i,1,:] = centroid_1

  n = feature.shape[0]
  loss = (1/n) * (np.sum(distance_0[label_feature_02 == 0]) + \
                  np.sum(distance_1[label_feature_02 == 1]))

  loss_iteration_02[i] = loss
  label_feature_02 = compute_label(distance_all)

# ++++++++++++++++++++++++++++++++++++++++++++++++++
```

## the number of clusters $K = 4$

```python
number_cluster          = 4
number_iteration        = 300     # you can modify this value
loss_iteration_04       = np.zeros(number_iteration)
centroid_iteration_04   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_04        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
for i in range(number_iteration): 

  centroid_0 = compute_centroid(feature, label_feature_04, 0) 
  centroid_1 = compute_centroid(feature, label_feature_04, 1) 
  centroid_2 = compute_centroid(feature, label_feature_04, 2) 
  centroid_3 = compute_centroid(feature, label_feature_04, 3) 

  distance_0 = compute_distance(feature, centroid_0).reshape(-1,1)
  distance_1 = compute_distance(feature, centroid_1).reshape(-1,1)
  distance_2 = compute_distance(feature, centroid_2).reshape(-1,1)
  distance_3 = compute_distance(feature, centroid_3).reshape(-1,1)

  distance_all = np.concatenate([distance_0, distance_1, 
                                 distance_2, distance_3], axis=1)
  
  centroid_iteration_04[i,0,:] = centroid_0
  centroid_iteration_04[i,1,:] = centroid_1
  centroid_iteration_04[i,2,:] = centroid_2
  centroid_iteration_04[i,3,:] = centroid_3

  n = feature.shape[0]
  loss = (1/n) * (np.sum(distance_0[label_feature_04 == 0]) + \
                  np.sum(distance_1[label_feature_04 == 1]) + \
                  np.sum(distance_2[label_feature_04 == 2]) + \
                  np.sum(distance_3[label_feature_04 == 3]))

  loss_iteration_04[i] = loss
  label_feature_04 = compute_label(distance_all)

#
# ++++++++++++++++++++++++++++++++++++++++++++++++++
```
